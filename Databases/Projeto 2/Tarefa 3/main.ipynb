```python
# Ajustando o price para ser float
df_calendar['price'] = df_calendar['price'].str.replace('$', '').str.replace(',', '').astype(float)
df_listings['price'] = df_listings['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Antes de criar as tabelas validar se os campos pensados podem ser chave
#calendar.csv
print(f" Tem listing_id e date repetido no csv calendar.csv: {df_calendar.duplicated(subset=['listing_id', 'date']).any()}")
#listings.csv
print(f" Tem id repetido no csv listings.csv:                {df_listings['id'].duplicated().any()}")
#reviews.csv
print(f" Tem id repetido no csv reviews.csv:                 {df_reviews['id'].duplicated().any()}")
```

## Criando as tabelas

```python
%%sql
CREATE TABLE IF NOT EXISTS Calendar (
    listing_id INT,
    date DATE,
    available BOOLEAN,
    price DECIMAL(10, 2),
    adjusted_price DECIMAL(10, 2),
    minimum_nights INT,
    maximum_nights INT,
    PRIMARY KEY (listing_id, date)
);
```

```python
%%sql

CREATE TABLE IF NOT EXISTS Reviews (
    id BIGINT PRIMARY KEY,
    listing_id BIGINT,
    date DATE,
    reviewer_id BIGINT,
    reviewer_name VARCHAR(255),
    comments TEXT
);
```

```python
%%sql
CREATE TABLE IF NOT EXISTS Listings (
    id INT PRIMARY KEY,
    name VARCHAR(255),
    host_id INT,
    host_name VARCHAR(255),
    neighbourhood_group VARCHAR(100),
    neighbourhood VARCHAR(100),
    latitude DECIMAL(10, 6),
    longitude DECIMAL(10, 6),
    room_type VARCHAR(100),
    price DECIMAL(10, 2),
    minimum_nights INT,
    number_of_reviews INT,
    last_review DATE,
    reviews_per_month DECIMAL(6, 2),
    calculated_host_listings_count INT,
    availability_365 INT,
    number_of_reviews_ltm INT,
    license VARCHAR(100)
);
```
