# **Exercício 1**
Considere o espaço vetorial real $\mathbb{R}^2$. Sejam as transformações
- $T_1:\mathbb{R}^2 \to \mathbb{R}^2$ tal que $T_1(x,y)=(-x,y)$ - **reflexão em torno do eixo $oy$**;
- $T_2:\mathbb{R}^2 \to \mathbb{R}^2$ tal que a matriz da transformação é dada por
$$[T_2] = \begin{bmatrix} 1 & k \\ 0 & 1 \end{bmatrix},$$
com $k\in\mathbb{R}$ - **cisalhamento horizontal**.

1. Aplique as transformações $T_1$ e $T_2$ no quadrado de vértices $(0,0)$, $(1,0)$, $(1,1)$ e $(0,1)$ considerando o parâmetro $k= -0,5$ e visualize os resultados;
2. Determine a matriz da transformação $T_3$ que primeiro faz um cisalhamento horizontal com $k=-0.5$ mapeando $e_2$ em $e_2 - 0.5e_1$ ($e_1$ se mantém inalterado) e então reflete o resultado em torno do eixo $oy$. **Dica:** Determine a posição final das imagens de $e_1$ e $e_2$.
3. Qual a relação entre as matrizes $[T_1]$, $[T_2]$ e a matriz da transformação $[T_3]$ obtida no item anterior?
4. Determine a transformação de reflexão em torno do eixo $ox$ e a transformação de cisalhamento vertical.

```python
# Solução 1

# Vértices do quadrado
Q = np.array([[0,0],[1,0],[1,1],[0,1]]).T

# Reflexão em torno do eixo y
# (-x,y) = x(-1,0) + y(0,1)
T1 = np.array([[-1,0],[0,1]]).T 

# Cisalhamento horizontal
T2 = np.array([[1,-0.5],[0,1]])

## Aplicando as transformações
# aplicando a reflexão em torno do eixo y no quadrado Q
T1Q = T1 @ Q
# aplicando o cisalhamento horizontal no quadrado Q
T2Q = T2 @ Q

def plotSimples(vx, vy):
  plt.figure(figsize=(6,6))
  plt.axis([-2, 2, -2, 2])
  plt.plot([-2,2], [0,0], 'k--', alpha=0.75)
  plt.plot([0,0], [-2,2],'k--', alpha=0.75)
  plt.plot(vx, vy, 'k')
  plt.plot([vx[0], vx[-1]], [vy[0], vy[-1]], 'k')
  plt.plot(vx, vy, 'bo')
  plt.grid('True')

# Plots
plotSimples(Q[0,:], Q[1,:])
plt.title('Vértices originais')

plotSimples(T1Q[0,:], T1Q[1,:])
plt.title('Reflexão')

plotSimples(T2Q[0,:], T2Q[1,:])
plt.title('Cisalhamento')
```

```python
#Solução 2)

# Observando as posições finais de e1 e e2 temos:
# T3(1,0) = (-1,0)
# T3(0,1) = (0.5,1)
T3 =  np.array([[-1,0],[0.5,1]]).T
print('Matriz de transformação T3:')
print(T3)

```

```python
# Solução 3)

# Por outro lado, aplicando T2, depois T1
T21 = T1 @ T2Q

# Aplicando T3 em Q
T3Q = T3 @ Q

# multiplicação da matriz T1 pela matriz T2
T1T2 = T1 @ T2
print('Matriz resultante da multiplicação de T1 por T2:')
print(T1T2)

#Plots
plotSimples(Q[0,:], Q[1,:])
plt.title('Vértices originais')

plotSimples(T21[0,:], T21[1,:])
plt.title('Aplicando T2 em Q depois T1 em T2Q')

plotSimples(T3Q[0,:], T3Q[1,:])
plt.title('Aplicando T3 em Q')

# Observando o gráfico da transformação T3 em Q, e o gráfico de
# T2 em Q e depois T1, percebe-se que são idênticos, e ao multiplicar T1 por T2
# percebe-se que a matriz resultante é igual a T3, portante pode-se concluir
# que a relação entre T1, T2 e T3 é a seguinte: T3 = T1 @ T2


```

# **Exercício 3**

a) Dada uma base qualquer de um subespaço vetorial do $R^{n}$, escreva um código para encontrar uma base ortonormal para este subespaço. Teste o seu código para a base dada pelas colunas da matriz  **V**  gerada no código abaixo. 

b) Faça um teste para verificar que a base obtida é de fato ortonormal.

```python
V = np.random.randint(0,20,size=(20,10))
print(V)
```

```python
## a) Ortonormalização de uma base qualquer dada e teste do código em V

# foi utilizado o método de Gram Schmidt que é um método para
# ortonormalização de um conjunto de vetores
def gram_schmidt_classico(X):
  # faz uma cópia da base passada no parâmetro(X)
  V = X.copy()
  # cria uma matriz Q de zeros do mesmo tamanho de X
  Q = np.zeros(X.shape)
  # para j de 0 até o número de colunas de X
  for j in range(X.shape[1]):
    # salva em vj os valores de cada linha para a coluna j
    vj = V[:,j]
    # para k de 0 até j(coluna)
    for k in range(j):
      # processo de ortogonalização de cada vetor
      vj = vj - np.dot(X[:,j], Q[:,k]) * Q[:,k]
    # processo de ortonormalização de cada vetor
    qj = vj / np.linalg.norm(vj)
    # salva em Q os valores ortonormalizados
    Q[:,j] = qj
  # retorna a base ortonormal
  return Q

# encontrar uma base ortonormal para a base dada 
# pelas colunas da matriz V gerada randomicamente 
Q = gram_schmidt_classico(V)
print("Base encontrada: ")
print(Q)
print('\n\n')


## b) Teste que verifica se a base obtida é de fato ortonormal

# para verificar se a base Q é ortonormal, podemos fazer Q.T @ Q,
# se a multiplicação de Q pela sua transposta for igual a Identidade, 
# então a base é ortonormal
X = Q.T @ Q
print('Q.T @ Q: ')
print(X)

# Observando X, percebe-se que os elementos da diagonal principal são iguais a 1
# e o resto dos elementos iguais a 0 (ou muito próximo de 0, por conta da precisão 
# de ponto flutuante, já explicada anteriormente), logo a base encontrada é ortonormal.



```

Tabela: *População do Brasil, em milhões de pessoas, entre os anos de 1872-2010. Dados Censo-IBGE.*



O último Censo foi realizado em 2010 e, devido à pandemia da Covid-19, o de 2020 foi adiado. Por isso, não é possível saber, com maior precisão, a população atual do país. Com o objetivo de estimar a população do Brasil em 2022, ajuste, no sentido dos mínimos quadrados, uma reta e uma parábola aos dados representados na tabela.

a) Qual das duas aproximações você considera melhor para estimar a população atual do país ? Justifique a sua resposta calculando o erro da aproximação. Mostre, também, os gráficos das duas aproximações obtidas.

b) Qual seria a população estimada do Brasil em 2022 ? Justifique.

```python
# vetor dos anos realizados o Censo - IBGE
x = np.array([1872, 1890, 1900, 1920, 1940, 1950, 1960, 1970, 1980, 1991, 2000, 2010])
# vetor da população em milhões para os anos realizado o Censo 
y = np.array([9.9, 14.3, 17.4, 30.6, 41.2, 51.9, 70.9, 94.5, 121.1, 146.9, 169.5, 190.7])

## Cálculo dos Mínimos Quadrados
# Resolvendo a equação X^T.X a = X^T y (X matriz de Vandermond)
def min_squares(x, y, m):
  # gera uma matriz Vandermond a partir no vetor x (anos)
  A = np.vander(x, m, increasing=True) 
  # calcula A^T.A da equação, com a matriz de Vandermond(A)
  AtA = np.dot(A.T, A)
  # calcula A^T.y da equação, com a matriz de Vandermond(A) e o vetor y(população)
  Aty = np.dot(A.T, y)
  # resolve o sistema linear
  alpha = np.linalg.solve(AtA, Aty)
  # retorna os coeficientes 
  return alpha 


## Calculando o erro da aproximação
def min_squares_error(x, y, p):
  # salva os valores calculados pelo polinômio "p" para os valores do vetor x(anos)
  z = p(x[:])
  # o erro de aproximação, chamado erro médio quadrático,
  # é calculado pelo quadrado da distância de f a sua 
  # aproximação f', dividido pela quantidade de valores
  erro = np.linalg.norm(z - y[:])**2 / x.shape[0]
  return erro

alfa_reta = min_squares(x, y, 2) # calcula o sistema com m=2 (reta)
# Definindo P1(x) = a1*x + a0
p_reta = lambda x: alfa_reta[1] * x + alfa_reta[0]
print('Erro de aproximação para a reta: {}'.format(min_squares_error(x, y, p_reta))) # erro de aproximação
print(f'P_1(x) = {alfa_reta[1]}*x {alfa_reta[0]}\n') # imprime P1(x)

alfa_parabola = min_squares(x, y, 3) # calcula o sistema com m=3 (parabola)
# Definindo P2(x) = a2*x^2 + a1*x + a0
p_parabola = lambda x: alfa_parabola[2] * x**2 + alfa_parabola[1] * x + alfa_parabola[0]
print('Erro de aproximação para a parabola: {}'.format(min_squares_error(x, y, p_parabola))) # erro de aproximação
print(f'P_2(x) = {alfa_parabola[2]}*x\u00b2 {alfa_parabola[1]}*x + {alfa_parabola[0]}\n') # imprime P2(x)

## Estimativa da população do Brasil em 2022
# como analisado, o ajuste feito pela parábola 
# estima melhor a população, portanto utilizaremos ele para estimar
pop_estimada = p_parabola(2022)

x_dist = np.linspace(1872, 2010, num=100)

plt.scatter(x,y,label = 'amostra',linewidth = 1,color='red')
plt.plot(x_dist,p_reta(x_dist),label = 'aproximação por reta',linewidth = 2)
plt.plot(x_dist,p_parabola(x_dist),label = 'aproximação por parábola',linewidth = 2)
plt.xlabel('Anos desde 1872',fontsize='large')
plt.ylabel('População (milhões)',fontsize='large')
plt.legend()
plt.show()

print()
# imprime a população estimada
print('População estimada do Brasil em 2022: {}\n'.format(pop_estimada)) 
```
