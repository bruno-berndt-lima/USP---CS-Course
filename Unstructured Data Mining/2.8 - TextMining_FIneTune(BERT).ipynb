# Dataset

```python
!wget https://raw.githubusercontent.com/rmarcacini/text-collections/master/complete_texts_csvs/Dmoz-Health.csv
```

```python
import pandas as pd

dataset = pd.read_csv('Dmoz-Health.csv')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(dataset, test_size=0.33, random_state=42)
```

```python
df_train
```

```python
df_test
```

```python
# Filtrando o df_train para as classes Senior e Mental
df_train = df_train[df_train['class'].isin(['Senior', 'Mental'])]
df_train['class'] = df_train['class'].map({'Senior': 0, 'Mental': 1})

# Filtrando o df_test para as classes Senior e Mental
df_test = df_test[df_test['class'].isin(['Senior', 'Mental'])]
df_test['class'] = df_test['class'].map({'Senior': 0, 'Mental': 1})
```

```python
df_train
```

# Ajuste Fino do BERT

```python
(x_train, y_train), (x_test, y_test), preproc = text.texts_from_df(df_train,
                                                                   'text', # coluna com dados textuais
                                                                   label_columns='class', # aqui pode ser multilabel
                                                                   maxlen=128,
                                                                   max_features=10000,
                                                                   preprocess_mode='bert',
                                                                   lang=None,
                                                                   val_pct = 0.1,
                                                                   random_state=42
                                                                   )
```

```python
model = text.text_classifier('bert', (x_train, y_train) , preproc=preproc)
classifier = ktrain.get_learner(model,
                             train_data=(x_train, y_train),
                             val_data=(x_test, y_test),
                             batch_size=32
                             )
```

```python
model.summary()
```

```python
counter = 0
for layer in model.layers:
    model.layers[counter].trainable = False

model.layers[105].trainable=True
model.layers[106].trainable=True

for layer in model.layers: print(layer.trainable)
```

```python
model.summary()
```

```python
classifier.fit_onecycle(0.00001,10)
```

```python
predictor = ktrain.get_predictor(classifier.model, preproc)
```

```python
df_test.text.to_list()
```

```python
y_pred = predictor.predict(df_test.text.to_list())
y_pred
```

```python
y_true = []
for c in df_test['class'].to_list():
  if c==0: y_true.append('not_class')
  else: y_true.append('class')

y_true
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred))
```
