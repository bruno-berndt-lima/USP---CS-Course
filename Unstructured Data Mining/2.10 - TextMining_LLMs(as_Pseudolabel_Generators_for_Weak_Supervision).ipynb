```python
import pandas as pd

dataset = pd.read_csv('Dmoz-Health.csv')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(dataset, test_size=0.33, random_state=42)
```

```python
df_test
```

```python
# Filtrando o df_train para as classes Senior e Mental
df_train = df_train[df_train['class'].isin(['Senior', 'Mental'])]
#df_train['class'] = df_train['class'].map({'Senior': 0, 'Mental': 1})

# Filtrando o df_test para as classes Senior e Mental
df_test = df_test[df_test['class'].isin(['Senior', 'Mental'])]
#df_test['class'] = df_test['class'].map({'Senior': 0, 'Mental': 1})
```

```python
df_train
```

```python
df_train_rotulado, df_train_nao_rotulado = train_test_split(df_train, test_size=0.9, random_state=42)
df_train_rotulado
```

```python
df_train_rotulado.text.to_list()[2]
```

```python
from tqdm.notebook import tqdm
L_pseudolabeled = []

for index,row in tqdm(df_train_nao_rotulado.iterrows(),total=len(df_train_nao_rotulado)):
  instrucao = "Selecione um rótulo para o texto abaixo dentre as opções Mental e Senior, conforme os exemplos abaixo. Informe apenas um rótulo, sem explicação."

  exemplo1 = "@text = Ms. Sue Bowen, L.C.S.W., C.A.C. Healing and hope available on phone or email. A licensed clinician offers guidance for relationships and life stressors. Includes qualifications, and contact details."
  exemplo1 += "@label = Mental"

  exemplo2 = "@text = Spruce Point Assisted Living Community Spruce Point is a 47-apartment assisted living community for seniors in Florence, Oregon. "
  exemplo2 += "@label = Senior"

  novo_texto = "@text = "+row['text']
  novo_texto += "@label = "

  prompt = instrucao + "\n\n"+exemplo1+"\n\n"+exemplo2+"\n\n"+novo_texto
  #print(prompt)
  v = llm_query(prompt)
  #print("Saida: ",llm_query(prompt))
  if v[0]=='Mental' or v[0]=='Senior':
    v_dict = row.to_dict()
    v_dict['pseudo_label'] = v[0]
    L_pseudolabeled.append(v_dict)

df_pseudolabeled = pd.DataFrame(L_pseudolabeled)
```

```python
df_pseudolabeled
```

```python
from sklearn.metrics import classification_report
print(classification_report(df_pseudolabeled['class'], df_pseudolabeled['pseudo_label']))
```

```python
L_final = []

for index,row in df_train_rotulado.iterrows():
  v = row.to_dict()
  v['y_true'] = str(v['class'])
  L_final.append(v)

for index,row in df_pseudolabeled.iterrows():
  v = row.to_dict()
  v['y_true'] = str(v['pseudo_label'])
  L_final.append(v)

df_final = pd.DataFrame(L_final)
df_final
```

```python
!pip install -U sentence-transformers
from sentence_transformers import SentenceTransformer
import numpy as np
import logging

# Load Sentence model (based on BERT) from URL
model = SentenceTransformer('distiluse-base-multilingual-cased-v2')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3,metric="cosine")

X_train = model.encode(df_final.text.to_list())
X_test = model.encode(df_test.text.to_list())

knn.fit(X_train, df_final['y_true'])
y_pred = knn.predict(X_test)
y_pred
```

```python
print(classification_report(df_test['class'], y_pred))
```
