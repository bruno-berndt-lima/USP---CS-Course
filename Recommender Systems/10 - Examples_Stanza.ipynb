```python
import stanza
stanza.download('en') # download English model
```

```python
nlp = stanza.Pipeline('en', processors='tokenize,mwt,pos,sentiment')
```

```python
doc = nlp("Barack Obama was born in Hawaii. I am sad. Now, you are happy!")
```

```python
doc.sentences[0].words
```
